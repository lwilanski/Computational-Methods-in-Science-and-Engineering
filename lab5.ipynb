```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def draw_sphere(num_points_s, num_points_t):
    s = np.linspace(0, 2 * np.pi, num_points_s)
    t = np.linspace(0, np.pi, num_points_t)

    s, t = np.meshgrid(s, t)

    x = np.cos(s) * np.sin(t)
    y = np.sin(s) * np.sin(t)
    z = np.cos(t)

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, color='b', alpha=0.6)

    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    
    ax.set_title('3D Unit Sphere')

    plt.show()
```

```python
def draw_transformed_sphere(num_points_s, num_points_t, matrix):
    s = np.linspace(0, 2 * np.pi, num_points_s)
    t = np.linspace(0, np.pi, num_points_t)

    s, t = np.meshgrid(s, t)

    x = np.cos(s) * np.sin(t)
    y = np.sin(s) * np.sin(t)
    z = np.cos(t)

    transformed_points = matrix @ np.vstack((x.flatten(), y.flatten(), z.flatten()))

    x_transformed = transformed_points[0, :].reshape(num_points_t, num_points_s)
    y_transformed = transformed_points[1, :].reshape(num_points_t, num_points_s)
    z_transformed = transformed_points[2, :].reshape(num_points_t, num_points_s)

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x_transformed, y_transformed, z_transformed, color='b', alpha=0.6)

    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    ax.set_title('Transformed Sphere into Ellipsoid')

    plt.show()
```

```python
draw_sphere(50, 50)
```

```python
def generate_random_matrix(a, b):
    return np.random.uniform(a, b, (3, 3))
```

```python
m = generate_random_matrix(0, 1)
for row in m:
    print(row)
```

```python
draw_transformed_sphere(50, 50, m)
```
